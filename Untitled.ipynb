```python
import numpy as np
import matplotlib.pyplot as plt

# X-axis: model complexity
x = np.linspace(0, 10, 200)

# Adjusted example functions
bias2 = 10 * np.exp(-0.4 * x)          # starts higher, decays
variance = 0.3 * np.exp(0.4 * x)      # starts low, grows
irreducible_error = 1.0               # noise floor

test_error = bias2 + variance + irreducible_error

# Plotting
plt.figure(figsize=(8,6))
plt.plot(x, bias2, label="Bias²", linestyle="--", color="blue")
plt.plot(x, variance, label="Variance", linestyle="--", color="red")
plt.plot(x, test_error, label="Test Error", color="black", linewidth=2)
plt.axhline(irreducible_error, color="gray", linestyle=":", label="Irreducible Error")

# Labels only,
plt.xlabel("Model Complexity")
plt.ylabel("Error")
plt.title(" Classical Bias-Variance Tradeoff")
plt.legend()
plt.grid(True)
plt.show()
```
